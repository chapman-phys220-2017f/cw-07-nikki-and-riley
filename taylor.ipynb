$$f(a + y) = \sum_{n=0}^\infty \frac{y^n}{n!}\left[\frac{d^n}{dy^n}f(a+y)\right]_{y=0}\\ = f(a) + f'(a)y + f''(a)\frac{y^2}{2} + \cdots$$

_a= point on x axis, y= parameter on the x-axis describing area surrounding point a, n= number of terms in Taylor expansion sum_

#### Explaining Taylor's Theorem:

Taylor's Theorem applies to a function with point 'a'.  This theorem says that, if function f is infinitely differentiable, a polynomial f(a+y) can be created so that all of the ordered derivative values are going to equal the function around point a.  The 'y' in this formula refers to a parameter on the x-axis that describes the surrounding area around point a.  This enables the formula to approximate the function around point a.  This theorem is useful when one wants to determine the shape of a fuction specifically around one point, for instance in an Energy vs Bond Length graph.
